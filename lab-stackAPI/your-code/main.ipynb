# StackAPI

#### Import the necessary libraries here:

```python
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
```

```python
comments = SITE.fetch('comments')
comments
```

#### Question 1: Find the questions and answers of last month.

```python
questions = SITE.fetch('questions', fromdate=1559390400, todate=1561896000)
questions
```

```python
answers = SITE.fetch('answers', fromdate=1559390400, todate=1561896000)
answers
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# Define the dates using datetime
import datetime
a = datetime.datetime(2019, 7, 12)
b = datetime.datetime(2019, 7, 13)
```

```python
# Answer question 2
questions = SITE.fetch('questions', fromdate=a, todate=b, min=5)
questions
```

#### Question 3: Find the answers with id 6784 and 6473.

```python
# Answer question 3
answers_id = SITE.fetch('answers', ids=[6784, 6473])
answers_id
```
